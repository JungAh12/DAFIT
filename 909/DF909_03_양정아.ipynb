```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```
#Sharpening
#3x3 kernel
kernel = np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
img = cv2.imread('/content/drive/My Drive/다핏문제/909/lena.jpg',2)
img = cv2.resize(img,(256,256))
```

```
cv2_imshow(img)
```

```
img = cv2.filter2D(img,-1,kernel)
```
