```
import re

filePath = 'drive/My Drive/다핏문제/01/Beta01_NLP.txt'
p = re.compile('[ A-Z-a-z]')
f = open(filePath, mode='r', encoding='euc-kr')
s = f.read()
f.close()
s = p.findall(s)
print(s)
s = ''.join(s).split(' ')
s = list(set(s))
```

```
s = s[1:]
```

```
for i in range(len(s)):
    s[i] = s[i].lower()
s = sorted(s)
```

```
f = open('drive/My Drive/다핏문제/03/DF901_03_양정아.txt', mode='w', encoding='euc-kr')
f.write('\n'.join(s))
f.close()
```
