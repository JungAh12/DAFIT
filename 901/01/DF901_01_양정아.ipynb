```
import re

inputFile = 'Beta01_NLP.txt'
outputFile = 'DF901_01_양정아.txt'
p = re.compile('[ ㄱ-ㅣ가-힣A-Za-z]')
f = open(inputFile, mode='r', encoding='euc-kr')
s = f.read()
f.close()
s = p.findall(s)
s = ''.join(s)
print(s)
f = open(outputFile, mode='w', encoding='euc-kr')
f.write(s)
f.close()
```
