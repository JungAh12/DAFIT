```
#908_02
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
import openpyxl

nltk.download('punkt')
nltk.download('stopwords')

wb = openpyxl.load_workbook('/content/drive/My Drive/다핏문제/908/DF908_01.xlsx')
ws = wb.active
index = []
for r in ws.rows:
    index.append(r[2].value)

tokens = []

for idx in index:
    tokens.append(sent_tokenize(idx))

stop_words = set(stopwords.words('english'))

# result = []
# for t in word_token:
#     res = []
#     for word in t:
#         if word not in stop_words:
#             res.append(word)
#     result.append(res)
```

```
sen_word_token=[]
word_token=[]
result = []
res = []
for sen in tokens:    
    for s in sen:
        word_token.append(word_tokenize(s))
        for words in word_token:
            for w in words:
                if w not in stop_words:
                    res.append(w)
            result.append(res)
            res.clear()
        sen_word_token.append(result)
        result.clear()
```
