불용어란 분석에 있어서 큰 의미가 없는 단어들을 말한다.

```
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
import openpyxl

nltk.download('punkt')
nltk.download('stopwords')
```

```
wb = openpyxl.load_workbook('/content/drive/My Drive/다핏문제/908/DF908_01.xlsx')
ws = wb.active
index = []
for r in ws.rows:
    index.append(r[2].value)
```

```
index[1]
```

```
tokens = []

for idx in index:
    tokens.append(word_tokenize(idx))
```

```
stop_words = set(stopwords.words('english'))
```

```
result = []
for t in tokens:
    res = []
    for word in t:
        if word not in stop_words:
            res.append(word)
    result.append(res)
```
